```python
# Open the file
with open('train_predictions.txt', 'r') as file:
    lines = file.readlines()

# Extract the text between <pad> and </s>
extracted_lines = [line[line.find('<pad>') + 5 : line.find('</s>')].strip() for line in lines]

# Write the extracted text to a new file
with open('formatted_predictions.txt', 'w') as file:
    for line in extracted_lines:
        file.write(line + '\n')
```
